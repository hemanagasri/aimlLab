```python
import numpy as np

# Example data
#data = [12, 16, 20, 24, 28, 32, 36, 40, 44, 48]

# Calculating quartiles
q1 = np.percentile(data['mpg'], 25)
q2 = np.percentile(data['mpg'], 50)
q3 = np.percentile(data['mpg'], 75)

print("First Quartile (Q1):", q1)
print("Second Quartile (Q2 - Median):", q2)
print("Third Quartile (Q3):", q3)
print("iqr ",q3-q1)
```

```python
# Using Seaborn
# Assuming your dataset has multiple columns (attributes
df=data.iloc[:,:-1]
print(df.columns)
for column in df.columns:
    plt.figure()  # Create a new figure for each attribute
    #df.boxplot(column=column)
    plt.boxplot(df[column], showfliers=True)
    plt.title(f'Boxplot of {column}')
    plt.ylabel(column)
    plt.show()
```

```python
for attribute in df.columns:
  Q1 = df[attribute].quantile(0.25)
  Q3 = df[attribute].quantile(0.75)
  IQR = Q3 - Q1
  lower_bound = Q1 - 1.5 * IQR
  upper_bound = Q3 + 1.5 * IQR
  print("lower_bound",lower_bound)
  print("upper_bound",upper_bound)
  outliers = df[(df[attribute] < lower_bound) | (df[attribute] > upper_bound)]
  print("Outliers:")
  print(outliers[[attribute]].count())
  print(outliers[[attribute]])
  # Cap the outliers below the lower bound to the lower bound value
  df.loc[df[attribute] < lower_bound, attribute] = lower_bound

  # Cap the outliers above the upper bound to the upper bound value
  df.loc[df[attribute] > upper_bound, attribute] = upper_bound
```
